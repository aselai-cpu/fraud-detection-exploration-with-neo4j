```python
import random as rand
nums =[i for i in range(11)]
print(nums)

prob = [ round( i / sum(nums),3) for i in nums ]
print(prob)
print(f"{sum(prob)} : max = {max(nums)} : min {min(nums)}")

for _ in range(10):
    print(rand.choice(nums), end= " ")
print()
for _ in range(10):
    print(rand.choices(nums,prob), end= " ")

```
