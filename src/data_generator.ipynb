```python
from faker import Faker

#Faker.__dict__
#help(Faker)

faker = Faker()

print(faker.first_name())
#faker.device_name()
faker.city()

#print(faker.__attrs__)
print(faker.email())
print(faker.url())
print(faker.profile())
```

```python
import pandas as pd

faker.name()
```

```python

data = pd.DataFrame()

for i in range(100):
    name = faker.name()
    data.loc[i, 'name'] = name
    data.loc[i, 'length'] = len(name)

print(data)

```

```python
from  matplotlib import pyplot

pyplot.bar(data['name'], data['length'])
```

```python
pyplot.scatter(data['name'], data['length'])
```
